```python
import pandas as pd
import time

df_pd = pd.read_csv('pothole_complaints.csv')

start = time.time()

# A realistic cleaning + aggregation pipeline
df_pd['created_date'] = pd.to_datetime(df_pd['created_date'])
df_pd_clean = df_pd.dropna(subset=['latitude', 'longitude'])
risk_scores = df_pd_clean.groupby('borough').agg(
    complaint_count=('complaint_type', 'count'),
    avg_days_open=('days_open', 'mean')
).reset_index()
risk_scores['risk_score'] = risk_scores['complaint_count'] * risk_scores['avg_days_open']
risk_scores = risk_scores.sort_values('risk_score', ascending=False)

pandas_time = time.time() - start
print(f"Pandas (CPU) time: {pandas_time:.4f} seconds")
print(risk_scores)
```

```python
import cudf
import time

df_gpu = cudf.read_csv('pothole_complaints.csv')

start = time.time()

df_gpu['created_date'] = cudf.to_datetime(df_gpu['created_date'].str.slice(0, 19))
df_gpu_clean = df_gpu.dropna(subset=['latitude', 'longitude'])
risk_scores_gpu = df_gpu_clean.groupby('borough').agg(
    complaint_count=('complaint_type', 'count'),
    avg_days_open=('days_open', 'mean')
).reset_index()
risk_scores_gpu['risk_score'] = risk_scores_gpu['complaint_count'] * risk_scores_gpu['avg_days_open']
risk_scores_gpu = risk_scores_gpu.sort_values('risk_score', ascending=False)

cudf_time = time.time() - start
print(f"cuDF (GPU) time: {cudf_time:.4f} seconds")
print(risk_scores_gpu)
```

```python
speedup = pandas_time / cudf_time
print(f"Pandas (CPU) time: {pandas_time:.4f} seconds")
print(f"cuDF (GPU) time: {cudf_time:.4f} seconds")
print(f"Speed-up: {speedup:.2f}x faster with GPU acceleration")
```

```python
import matplotlib.pyplot as plt

methods = ['Pandas (CPU)', 'cuDF (GPU)']
times = [pandas_time, cudf_time]

plt.figure(figsize=(6,4))
plt.bar(methods, times, color=['#4285F4', '#34A853'])
plt.ylabel('Time (seconds)')
plt.title(f'Processing Time: {speedup:.1f}x Speed-up with GPU Acceleration')
plt.savefig('speedup_chart.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
from google.colab import auth
auth.authenticate_user()

from google.cloud import bigquery
import pandas as pd

project_id = "repairrank-hackathon"
client = bigquery.Client(project=project_id)
```

```python
query_big = """
SELECT
  complaint_type, borough, created_date, closed_date,
  latitude, longitude, status,
  DATE_DIFF(DATE('2022-07-30'), DATE(created_date), DAY) AS days_open
FROM `bigquery-public-data.new_york.311_service_requests`
WHERE complaint_type IN (
  'Street Condition', 'Noise - Residential', 'Illegal Parking',
  'Blocked Driveway', 'HEAT/HOT WATER', 'Request Large Bulky Item Collection'
)
"""
df_big = client.query(query_big).to_dataframe()
print(df_big.shape)
df_big.to_csv('big_complaints.csv', index=False)
```

```python
import cudf
df_gpu = cudf.read_csv('big_complaints.csv')
```

```python
import time

start = time.time()
df_gpu['created_date'] = cudf.to_datetime(df_gpu['created_date'].str.slice(0, 19))
df_gpu_clean = df_gpu.dropna(subset=['latitude', 'longitude'])
risk_scores_gpu = df_gpu_clean.groupby('borough').agg(
    complaint_count=('complaint_type', 'count'),
    avg_days_open=('days_open', 'mean')
).reset_index()
risk_scores_gpu['risk_score'] = risk_scores_gpu['complaint_count'] * risk_scores_gpu['avg_days_open']
risk_scores_gpu = risk_scores_gpu.sort_values('risk_score', ascending=False)
cudf_time = time.time() - start
print(f"cuDF (GPU) time: {cudf_time:.4f} seconds")
print(risk_scores_gpu)

# Save immediately so we don't lose this again if runtime disconnects
risk_scores_gpu.to_pandas().to_csv('risk_scores.csv', index=False)
```

```python
top_boroughs = risk_scores_gpu.to_pandas()

summary_text = ""
for _, row in top_boroughs.iterrows():
    summary_text += f"- {row['borough']}: {row['complaint_count']} street condition complaints, average {row['avg_days_open']:.1f} days open, risk score {row['risk_score']:.0f}\n"

print(summary_text)
```

```python
prompt = f"""
You are a city infrastructure advisor. Based on the following street condition complaint data by borough in NYC, write a brief, actionable repair-prioritization recommendation for city officials. Be specific about which borough needs urgent attention and why, using the data provided.

Data:
{summary_text}

Write 3-4 sentences, practical and clear, suitable for a city council briefing.
"""

response = client_ai.models.generate_content(
    model='gemini-2.5-flash',
    contents=prompt
)
print(response.text)
```

```python
with open('ai_recommendation.txt', 'w') as f:
    f.write(response.text)

print("Saved!")
print(response.text)  # print full text to see the complete recommendation
```
